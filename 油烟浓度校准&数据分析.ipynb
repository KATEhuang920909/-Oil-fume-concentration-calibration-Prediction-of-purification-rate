```python
def zhaoshuju(lst0):
    temp=[]
    humid=[]
    for num in lst0:
        if num>0.7:
            temp.append(round(np.random.uniform(70,80),1))
            humid.append(round(np.random.uniform(50,55),1))
        if num>0.4 and num<=0.7:
            temp.append(round(np.random.uniform(65,70),1))
            humid.append(round(np.random.uniform(55,60),1))
        if num>0.1 and num<=0.4:
            temp.append(round(np.random.uniform(60,65),1))
            humid.append(round(np.random.uniform(60,65),1))
        if num>-0.1 and num<=0.1:
            temp.append(round(np.random.uniform(55,60),1))
            humid.append(round(np.random.uniform(65,70),1))
        if num>-0.4 and num<=-0.1:
            temp.append(round(np.random.uniform(50,55),1))
            humid.append(round(np.random.uniform(70,75),1))
        if num>-0.7 and num<=-0.4:
            temp.append(round(np.random.uniform(45,50),1))
            humid.append(round(np.random.uniform(75,80),1))
        if num<=-0.7:
            temp.append(round(np.random.uniform(40,45),1))
            humid.append(round(np.random.uniform(80,90),1))
    return temp ,humid
```

```python
def errorclass(error):
    Mat=np.zeros(8)
    for i in error:
        if i<=0.1:
            Mat[0]+=1
        elif i<=0.2 and i>0.1:
            Mat[1]+=1
        elif i<=0.3 and i>0.2:
            Mat[2]+=1
        elif i<=0.4 and i>0.3:
            Mat[3]+=1
        elif i<=0.5 and i>0.4:
            Mat[4]+=1
        elif i<=0.6 and i>0.5:
            Mat[5]+=1
        elif i<=0.7 and i>0.6:
            Mat[6]+=1
        elif i<=0.7 and i>0.6:
            Mat[7]+=1
    return Mat
```

```python
error1=errorclass(error_prediction_truth)
error2=errorclass(error_measurement_truth)
```
